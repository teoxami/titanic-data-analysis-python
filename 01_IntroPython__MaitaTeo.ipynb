## 📊 Lectura del Dataset
Se carga el dataset en un DataFrame de pandas para poder manipular y analizar la información.


```python
# Importar librerías necesarias
import pandas as pd
import numpy as np

# Cargar el CSV a un DataFrame
df = pd.read_csv('titanic.csv')

# Ver las primeras filas para confirmar
print("Primeras 5 filas del dataset:")
df.head()
```

## 📈 Análisis Estadístico General
Se utilizan estadísticas descriptivas para comprender la distribución y características principales de los datos.


```python
# Estadísticas descriptivas de los datos numéricos
print("Estadísticas descriptivas:")
df.describe()
```

## 🛟 Análisis de Supervivencia
En esta sección se calcula el porcentaje de pasajeros que sobrevivieron al accidente del Titanic.


```python
# BLOQUE 1: Análisis de supervivencia
print("="*50)
print("ANÁLISIS DE SUPERVIVENCIA")
print("="*50)

# Calcular porcentaje de supervivencia
supervivencia = df['Survived'].value_counts(normalize=True) * 100

print(f"✓ Porcentaje que SOBREVIVIÓ: {supervivencia[1]:.2f}%")
print(f"✗ Porcentaje que NO sobrevivió: {supervivencia[0]:.2f}%")
print(f"\n Total de pasajeros: {len(df)} personas")
print(f"   Sobrevivieron: {df['Survived'].sum()} personas")
print(f"   Fallecieron: {len(df) - df['Survived'].sum()} personas")
```

## 🎟️ Análisis por Clase
Se analiza la distribución de los pasajeros según su clase (primera, segunda y tercera clase).


```python
# BLOQUE 2: Análisis de clase (Pclass)
print("\n" + "="*50)
print("ANÁLISIS POR CLASE (Pclass)")
print("="*50)

# Calcular porcentaje de pasajeros por cada clase
porcentaje_clase = df['Pclass'].value_counts(normalize=True) * 100
conteo_clase = df['Pclass'].value_counts()

# Mostrar resultados ordenados por clase
for clase in sorted(df['Pclass'].unique()):
    print(f"Clase {clase}: {porcentaje_clase[clase]:.2f}% del total ({conteo_clase[clase]} pasajeros)")

# Información adicional: supervivencia por clase
print("\n Supervivencia por clase:")
for clase in sorted(df['Pclass'].unique()):
    sobreviven_clase = df[df['Pclass'] == clase]['Survived'].mean() * 100
    print(f"   Clase {clase}: {sobreviven_clase:.2f}% de supervivencia")
```

## 🚻 Análisis por Género
Se determina la cantidad de hombres y mujeres que viajaban en el Titanic.


```python
# BLOQUE 3: Análisis por género
print("\n" + "="*50)
print("ANÁLISIS POR GÉNERO")
print("="*50)

# Contar hombres y mujeres
conteo_genero = df['Sex'].value_counts()
porcentaje_genero = df['Sex'].value_counts(normalize=True) * 100

print(f" Hombres: {conteo_genero.get('male', 0)} ({porcentaje_genero.get('male', 0):.2f}%)")
print(f" Mujeres: {conteo_genero.get('female', 0)} ({porcentaje_genero.get('female', 0):.2f}%)")

# Información adicional: supervivencia por género
print("\n Supervivencia por género:")
mujeres_superviven = df[df['Sex'] == 'female']['Survived'].mean() * 100
hombres_superviven = df[df['Sex'] == 'male']['Survived'].mean() * 100
print(f"    Mujeres: {mujeres_superviven:.2f}% sobrevivieron")
print(f"    Hombres: {hombres_superviven:.2f}% sobrevivieron")
```
